```python
import pandas as pd
from datetime import datetime

import plotly.graph_objects as go
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output, State
import plotly.express as px
```

```python
# load data
df = pd.read_csv("covid_country_data_perc.csv", index_col=0, na_filter=False, parse_dates=['Year-Month'])
df
```

```python
map_data = df.copy()
map_data = map_data.groupby(['Country/Region'])[["Confirmed_Case"]].max()

map_data.index.name = "Country/Region"
map_data.reset_index(inplace=True)
map_data = map_data.rename(columns={'Country/Region': 'Country'})
map_data
```

```python
app = dash.Dash()
```

```python
from urllib.request import urlopen
import json
with open('/Users/keke/Desktop/Data-Visualization/countries.geojson') as response:
    counties = json.load(response)
import plotly.graph_objects as go
```

```python
fig = px.choropleth(map_data, geojson=counties, locations='Country', color='Confirmed_Case',
                           color_continuous_scale="Viridis",
                           range_color=(0, 12),
                           scope="usa",
                           labels={'Confirmed_Case':'Confirmed Case'}
                          )
```

```python
fig = go.Figure(go.Choroplethmapbox(geojson=counties, locations=df.fips, z=df.unemp,
                                    colorscale="Viridis", zmin=0, zmax=12,
                                    marker_opacity=0.5, marker_line_width=0))
fig.update_layout(mapbox_style="carto-positron",
                  mapbox_zoom=3, mapbox_center = {"lat": 37.0902, "lon": -95.7129})
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
